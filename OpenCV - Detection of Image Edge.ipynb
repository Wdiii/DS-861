#### Loading gray image:

```python
image = cv2.imread('test.jpg') 

flower_image = np.copy(image)
gray =cv2.cvtColor(flower_image, cv2.COLOR_RGB2GRAY)

cv2.imshow('result', gray) 

cv2.waitKey(0) 
```

### Covert image to Grayscale and Reduce Noise

We have different wayt to do Image Blurring (Image Smoothing).

1) Averaging : blur = cv.blur(img,(5,5))

2) Gaussian Blurring : blur = cv.GaussianBlur(img,(5,5),0)  -- > we use this one in this notebook

3) Median Blurring : median = cv.medianBlur(img,5)

4) Bilateral Filtering : blur = cv.bilateralFilter(img,9,75,75)

https://docs.opencv.org/master/d4/d13/tutorial_py_filtering.html

```python
gray =cv2.cvtColor(flower_image, cv2.COLOR_RGB2GRAY)

#this step is optional when using Canny method
#reduce Noise from our grayscale image
blur = cv2.GaussianBlur(gray, (5,5), 0) 

cv2.imshow('result', blur) 
cv2.waitKey(0) 
```

### Canny Method: Edge Detection

A function of pixel intensities, in all directions X and y. - Derivative(f(x, y))
- small change, small derivative,
- large change, large derivative.

- cv2.Canny(image, low_threshold, high_threshold)

```python
# apply canny in blur image
canny = cv2.Canny(blur, 50, 150) #low threshold, high threshold

cv2.imshow('result', canny) 
cv2.waitKey(0) 
```
